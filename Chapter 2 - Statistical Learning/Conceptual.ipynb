# 1. For each of parts (a) through (d), indicate whether we would generally expect the performance of a flexible statistical learning method to be better or worse than an inflexible method. Justify your answer.

### (a) The sample size n is extremely large, and the number of predictors p is small.

Ans) As the data has few dimensions, the risk of a flexible method capturing noise is very low i.e. overfitting is avoided. As we have a large sample size, flexible methods will capture the complexity of the data without overfitting as both bias and variance will be low. Hence, using a flexible method is better.

### (b) The number of predictors p is extremely large, and the number of observations n is small.

Ans) For a large number of features, there is a high risk of overfitting as noise will be captured. Small sample size will not allow us to capture non-linear relationships and variance will be high as well. Additionally, we know that sample size needs to be exponentially larger as feature size increases. Here, we are presented with the opposite case, so we should opt for less flexible models.

### (c) The relationship between the predictors and response is highly non-linear.

Ans) Flexible models are better at capturing non-linear relationships.

### (d) The variance of the error terms, i.e. σ2 = Var(ϵ), is extremely high.

Ans) With high variance for errors, overfitting is likely to occur for flexible models. Hence, we should prefer non-flexible models.

# 3. We now revisit the bias-variance decomposition.

### (a) Provide a sketch of typical (squared) bias, variance, training error, test error, and Bayes (or irreducible) error curves, on a single plot, as we go from less flexible statistical learning methods towards more flexible approaches. The x-axis should represent the amount of flexibility in the method, and the y-axis should represent the values for each curve. There should be five curves. Make sure to label each one.

```python
import numpy as np
import matplotlib.pyplot as plt

# Define flexibility levels
flexibility = np.linspace(0, 1, 100)

# Generate hypothetical curves
bayes_error = 0.2 * np.ones_like(flexibility)  # Constant Bayes error
bias = 1 - np.power(flexibility, 1/5)
variance = 9/4 * flexibility**2 - 3/2 * flexibility + 1/4
training_error = np.power(bias, 3/2) * 4/5 + bayes_error   # Assuming equal contribution
test_error = (np.sqrt(bias) + variance) * 1/2 + bayes_error # Assumed to increase slightly

# Plot the curves
plt.plot(flexibility, bias, label='Bias', linestyle='--')
plt.plot(flexibility, variance, label='Variance', linestyle='--')
plt.plot(flexibility, training_error, label='Training Error')
plt.plot(flexibility, test_error, label='Test Error')
plt.plot(flexibility, bayes_error, label='Bayes Error', linestyle='--')

# Add labels and legend
plt.xlabel('Flexibility')
plt.ylabel('Error')
plt.yticks(ticks = [], labels=[])
plt.title('Bias-Variance Tradeoff')
plt.legend()
plt.show()
```

### (b) Explain why each of the five curves has the shape displayed in part (a).

- Bias : Reduces with flexibility
- Variance : Increases with flexibility due to overfitting
- Bayes Error : Constant and minimum possible
- Training Error : Reduces along with Bias due to overfitting but doesn't go below Bayes Error
- Test Error : Reduces along with Bias initially, but increases again wit Variance due to overfitting

# 5. What are the advantages and disadvantages of a very flexible (versus a less flexible) approach for regression or classification? Under what circumstances might a more flexible approach be preferred to a less flexible approach? When might a less flexible approach be preferred?

Ans)
Advantages:
1. For a large sample size and small set of features, a flexible model is able to capture the complexity of the data without overfitting
2. With a flexible model, we can capture non-linear and complex relationships
Disadvantages:
1. As we know that as size of feature set increases, the number of samples has to increase exponentially for there to be enought datapoints to capture the underlying pattern without overfitting. In cases with high-dimensional data or few data samples, flexible model will overfit to the training data, leading to high variance in the model
2. Interpretation is difficult for flexible models as the model has a large number of parameters
3. Computation time is high for more flexible models

Use more flexible models when:
1. Non-linear and complex relationships
2. large sample size
3. low feature set size
4. Computation resource availability not an issue
5. Interpretability not required

Use less flexible models when:
1. Small sample size
2. High feature set size
3. High correlation between predictors and target
4. Computation resources are low
5. Interpretation in needed

# Suppose we wish to use this data set to make a prediction for Y when X<sub>1</sub> = X<sub>2</sub> = X<sub>3</sub> = 0 using K-nearest neighbors.

### (a) Compute the Euclidean distance between each observation and the test point, X1 = X2 = X3 = 0.

Ans)

```python
import pandas as pd 
import numpy as np
from sklearn.metrics.pairwise import paired_euclidean_distances as ped

# Creating a dataframe in pandas
df = pd.DataFrame({'X1':[0,2,0,0,-1,1],'X2':[3,0,1,1,0,1],'X3':[0,0,3,2,1,1],'Y':['Red','Red','Red','Green','Green','Red']})
# Excluding target variable
Xes = df.iloc[:,:3]
# Computing euclidean distances
ped(Xes, np.zeros_like(Xes))
```

### (b) What is our prediction with K = 1? Why?

Ans) For k = 1, we will assign the same category as the nearest neighbor which is the 5th record, so Green

### (c) What is our prediction with K = 3? Why?

Ans) For k = 3, we select the same category as the highest frequency for the 3 nearest neighbors, so from 2nd, 5th, and 6th records, we get one green and two reds, so the answer is Red

### (d) If the Bayes decision boundary in this problem is highly non-linear, then would we expect the best value for K to be large or small? Why?

Ans) For a non-linear decision boundary, we would want a smaller value of k, as a small k will account for finer and local variations in the data. As we increase k, the decision boundary starts to generalize creating nearly linear boundaries.
